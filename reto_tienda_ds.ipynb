```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargamos los datos (ya están cargados en el notebook original)
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Añadimos una columna para identificar cada tienda
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

# Verificamos la estructura de los datos
print("Estructura de los datos de la Tienda 1:")
print(tienda1.info())
print("\nPrimeras filas de la Tienda 1:")
print(tienda1.head())

# 1. Análisis de facturación
print("\n\n--- 1. ANÁLISIS DE FACTURACIÓN ---")

# Calculamos la facturación total por tienda
facturacion_tienda1 = tienda1['Precio'].sum()
facturacion_tienda2 = tienda2['Precio'].sum()
facturacion_tienda3 = tienda3['Precio'].sum()
facturacion_tienda4 = tienda4['Precio'].sum()

print(f"Facturación total Tienda 1: ${facturacion_tienda1:,.2f}")
print(f"Facturación total Tienda 2: ${facturacion_tienda2:,.2f}")
print(f"Facturación total Tienda 3: ${facturacion_tienda3:,.2f}")
print(f"Facturación total Tienda 4: ${facturacion_tienda4:,.2f}")

# Creamos un DataFrame para visualizar la facturación
facturacion_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación': [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]
})

# Visualizamos la facturación por tienda
plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Facturación', data=facturacion_df)
plt.title('Facturación Total por Tienda')
plt.ylabel('Facturación ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 2. Ventas por categoría
print("\n\n--- 2. VENTAS POR CATEGORÍA ---")

# Función para analizar las categorías más vendidas por tienda
def analizar_categorias(tienda, nombre_tienda):
    ventas_por_categoria = tienda.groupby('Categoría del Producto').size().reset_index(name='Cantidad')
    ventas_por_categoria = ventas_por_categoria.sort_values('Cantidad', ascending=False)

    print(f"\nCategorías más vendidas en {nombre_tienda}:")
    print(ventas_por_categoria)

    # Visualizamos las categorías más vendidas
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Categoría del Producto', y='Cantidad', data=ventas_por_categoria)
    plt.title(f'Categorías más vendidas en {nombre_tienda}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return ventas_por_categoria

categorias_tienda1 = analizar_categorias(tienda1, 'Tienda 1')
categorias_tienda2 = analizar_categorias(tienda2, 'Tienda 2')
categorias_tienda3 = analizar_categorias(tienda3, 'Tienda 3')
categorias_tienda4 = analizar_categorias(tienda4, 'Tienda 4')

# 3. Calificación promedio de la tienda
print("\n\n--- 3. CALIFICACIÓN PROMEDIO DE LA TIENDA ---")

calificacion_tienda1 = tienda1['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

print(f"Calificación promedio Tienda 1: {calificacion_tienda1:.2f}/5")
print(f"Calificación promedio Tienda 2: {calificacion_tienda2:.2f}/5")
print(f"Calificación promedio Tienda 3: {calificacion_tienda3:.2f}/5")
print(f"Calificación promedio Tienda 4: {calificacion_tienda4:.2f}/5")

# Creamos un DataFrame para visualizar las calificaciones
calificaciones_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
})

# Visualizamos las calificaciones promedio
plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Calificación Promedio', data=calificaciones_df)
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación (1-5)')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# 4. Productos más y menos vendidos
print("\n\n--- 4. PRODUCTOS MÁS Y MENOS VENDIDOS ---")

# Función para analizar los productos más y menos vendidos por tienda
def analizar_productos(tienda, nombre_tienda):
    productos_count = tienda['Producto'].value_counts().reset_index()
    productos_count.columns = ['Producto', 'Cantidad']

    productos_mas_vendidos = productos_count.head(5)
    productos_menos_vendidos = productos_count.tail(5)

    print(f"\nProductos más vendidos en {nombre_tienda}:")
    print(productos_mas_vendidos)

    print(f"\nProductos menos vendidos en {nombre_tienda}:")
    print(productos_menos_vendidos)

    # Visualizamos los productos más vendidos
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Producto', y='Cantidad', data=productos_mas_vendidos)
    plt.title(f'Productos más vendidos en {nombre_tienda}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Visualizamos los productos menos vendidos
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Producto', y='Cantidad', data=productos_menos_vendidos)
    plt.title(f'Productos menos vendidos en {nombre_tienda}')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    return productos_mas_vendidos, productos_menos_vendidos

productos_tienda1 = analizar_productos(tienda1, 'Tienda 1')
productos_tienda2 = analizar_productos(tienda2, 'Tienda 2')
productos_tienda3 = analizar_productos(tienda3, 'Tienda 3')
productos_tienda4 = analizar_productos(tienda4, 'Tienda 4')

# 5. Envío promedio por tienda
print("\n\n--- 5. ENVÍO PROMEDIO POR TIENDA ---")

envio_tienda1 = tienda1['Costo de envío'].mean()
envio_tienda2 = tienda2['Costo de envío'].mean()
envio_tienda3 = tienda3['Costo de envío'].mean()
envio_tienda4 = tienda4['Costo de envío'].mean()

print(f"Costo de envío promedio Tienda 1: ${envio_tienda1:,.2f}")
print(f"Costo de envío promedio Tienda 2: ${envio_tienda2:,.2f}")
print(f"Costo de envío promedio Tienda 3: ${envio_tienda3:,.2f}")
print(f"Costo de envío promedio Tienda 4: ${envio_tienda4:,.2f}")

# Creamos un DataFrame para visualizar los costos de envío
envios_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de Envío Promedio': [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]
})

# Visualizamos los costos de envío promedio
plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Costo de Envío Promedio', data=envios_df)
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Resumen y recomendación final
print("\n\n--- RESUMEN Y RECOMENDACIÓN FINAL ---")

# Creamos un DataFrame con todos los indicadores
resumen_df = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturación Total': [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4],
    'Calificación Promedio': [calificacion_tienda1, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4],
    'Costo de Envío Promedio': [envio_tienda1, envio_tienda2, envio_tienda3, envio_tienda4]
})

print("Resumen de indicadores por tienda:")
print(resumen_df)

# Normalizamos los indicadores para poder compararlos
resumen_norm = resumen_df.copy()
resumen_norm['Facturación Normalizada'] = (resumen_df['Facturación Total'] - resumen_df['Facturación Total'].min()) / (resumen_df['Facturación Total'].max() - resumen_df['Facturación Total'].min())
resumen_norm['Calificación Normalizada'] = (resumen_df['Calificación Promedio'] - resumen_df['Calificación Promedio'].min()) / (resumen_df['Calificación Promedio'].max() - resumen_df['Calificación Promedio'].min())
# Para el costo de envío, menor es mejor, así que invertimos la normalización
resumen_norm['Envío Normalizado'] = 1 - ((resumen_df['Costo de Envío Promedio'] - resumen_df['Costo de Envío Promedio'].min()) / (resumen_df['Costo de Envío Promedio'].max() - resumen_df['Costo de Envío Promedio'].min()))

# Calculamos un score general (promedio de los indicadores normalizados)
resumen_norm['Score General'] = (resumen_norm['Facturación Normalizada'] + resumen_norm['Calificación Normalizada'] + resumen_norm['Envío Normalizado']) / 3

print("\nIndicadores normalizados y score general:")
print(resumen_norm[['Tienda', 'Facturación Normalizada', 'Calificación Normalizada', 'Envío Normalizado', 'Score General']])

# Identificamos la tienda con menor desempeño
tienda_menor_desempeno = resumen_norm.loc[resumen_norm['Score General'].idxmin(), 'Tienda']
print(f"\nLa tienda con menor desempeño general es: {tienda_menor_desempeno}")

# Visualizamos el score general
plt.figure(figsize=(10, 6))
sns.barplot(x='Tienda', y='Score General', data=resumen_norm)
plt.title('Score General por Tienda (Mayor es Mejor)')
plt.ylabel('Score (0-1)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualizamos todos los indicadores normalizados en un gráfico de radar
categorias = ['Facturación', 'Calificación', 'Envío']
fig = plt.figure(figsize=(10, 8))

# Creamos un gráfico de radar para cada tienda
for i, tienda in enumerate(resumen_norm['Tienda']):
    valores = [
        resumen_norm.loc[i, 'Facturación Normalizada'],
        resumen_norm.loc[i, 'Calificación Normalizada'],
        resumen_norm.loc[i, 'Envío Normalizado']
    ]

    # Cerramos el polígono repitiendo el primer valor
    valores.append(valores[0])

    # Calculamos los ángulos para cada categoría
    angulos = np.linspace(0, 2*np.pi, len(categorias), endpoint=False).tolist()
    angulos.append(angulos[0])  # Cerramos el círculo

    # Creamos el subplot
    ax = fig.add_subplot(2, 2, i+1, polar=True)
    ax.plot(angulos, valores, 'o-', linewidth=2, label=tienda)
    ax.fill(angulos, valores, alpha=0.25)
    ax.set_thetagrids(np.degrees(angulos[:-1]), categorias)
    ax.set_ylim(0, 1)
    ax.set_title(tienda)
    ax.grid(True)

plt.tight_layout()
plt.show()

# Conclusión final
print("\n--- CONCLUSIÓN FINAL ---")
print(f"Basado en el análisis de los datos, recomendamos al señor Juan vender la {tienda_menor_desempeno} para invertir en un nuevo negocio.")
print("Esta recomendación se basa en el análisis de facturación total, categorías más populares, calificación promedio de clientes,")
print("productos más y menos vendidos, y costo promedio de envío.")
```
